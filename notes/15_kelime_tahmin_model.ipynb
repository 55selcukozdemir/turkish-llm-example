```python
import torch
import torch.nn as nn

import torch.optim as optim

import numpy as np
```

```python
# Temel ayarlar
max_vocabilary_lenght = 70 # tr için uzun bunu buldum.
vocab_binary_lenght = 8 # 8 bit ile karakterleri tanımlamak istiyorum. Amacım boyut büyütmeden modeli oluşturabilmek. One-hot encoding de kullanılabilirdi.

feature_lenght = max_vocabilary_lenght * vocab_binary_lenght
```

```python
class KelimeTahmin(nn.Module):
    def __init__(self):
        super().__init__()

        self.linear1 = nn.Linear(feature_lenght, 2048)
        self.linear2 = nn.Linear(2048, 2048)

        self.relu = nn.ReLU()
        self.softmax = nn.Softmax()

    def forward(self, vocab):
        x = self.linear1(vocab)
        x = self.relu(x)
        x = self.linear2(x)
        x = self.softmax(x)

        return x
        
```

```python

# Model eğitimi
model = KelimeTahmin()
optimizer = optim.Adam(model.parameters())
criterion = nn.CrossEntropyLoss()

epochs = 20

for epoch in range(1, epochs):
    model.train()

    model()


```
