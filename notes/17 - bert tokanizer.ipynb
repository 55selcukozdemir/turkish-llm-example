```python
from transformers import BertTokenizer

# 1. BERT'in hazır tokenizer'ını yüklüyoruz
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# 2. Örnek cümlelerimiz (İster 1 cümle, ister bir liste dolusu cümle ver)
cumleler = [
    "Yapay zeka çok heyecan verici.",
    "BERT modelini eğitmek aslında çok kolay."
]

# 3. Cümleleri BERT'in anlayacağı formata çeviriyoruz
girdiler = tokenizer(
    cumleler, 
    padding=True,       # Cümle boylarını eşitlemek için kısa olanlara sıfır ekler
    truncation=True,    # Çok uzun cümleleri maksimum boyuta (örn: 512) keser
    max_length=16,      # Maksimum token uzunluğu sınırımız
    return_tensors="pt" # Çıktıyı PyTorch Tensor'ü olarak verir
)

# Bakalım arkada ne oluşmuş?
print("Kelimelerin Sayısal Karşılıkları (input_ids):\n", girdiler["input_ids"])
print("\nModelin Nereye Bakacağını Gösteren Maske (attention_mask):\n", girdiler["attention_mask"])
```
